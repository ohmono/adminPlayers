```python
def random_date(start, end):
    from random import randrange
    from datetime import timedelta, datetime
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)



for i in range(len(players)): #modify key 
    if players[i]['idType'] == 'TI':
        d1 = datetime.strptime('10/29/2002', '%m/%d/%Y')
        d2 = datetime.strptime('10/29/2008', '%m/%d/%Y')
        date = str(random_date(d1, d2)).split(' ')[0]

    elif players[i]['idType'] == 'CC':
        d1 = datetime.strptime('10/29/1999', '%m/%d/%Y')
        d2 = datetime.strptime('10/28/2002', '%m/%d/%Y')
        date = str(random_date(d1, d2)).split(' ')[0]

    players[i].update({'bornDate': date})
```
